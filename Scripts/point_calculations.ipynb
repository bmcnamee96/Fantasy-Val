```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import regex as re
import csv
```

```python
# Load the CSV file into a DataFrame
file_path = '../Data/games_data_updated.csv'
games_df = pd.read_csv(file_path)
print(f'{file_path} uploaded')
# Drop rows where both home_score and away_score are 0
games_df = games_df[~((games_df['home_score'] == 0) & (games_df['away_score'] == 0))]
games_df.head(5)
```

```python
# Load the CSV file into a DataFrame
file_path = '../Data/all_stats.csv'
stats_df = pd.read_csv(file_path)
print(f'{file_path} uploaded')
stats_df.tail(5)
```

```python
# Load the CSV file into a DataFrame
file_path = '../Data/series_df.csv'
series_df = pd.read_csv(file_path)
print(f'{file_path} uploaded')
series_df.tail(5)
```

```python
# Merge games_df into stats_df based on game_id
stats_df = pd.merge(stats_df, games_df[['game_id', 'series_id']], on='game_id', how='left')
```

```python
stats_df
```

```python
# Assuming you have your stats_df DataFrame

# Define points for each category
points_kills = 1
points_deaths = -0.5
points_assists = 0.5
points_adr = 0.1  
points_fk = 2
points_fd = -1
points_clutches = 2
points_aces = 3

# Calculate total points
stats_df['total_points'] = (
    stats_df['kills'] * points_kills +
    stats_df['deaths'] * points_deaths +
    stats_df['assists'] * points_assists +
    stats_df['adr'] * points_adr +
    stats_df['fk'] * points_fk +
    stats_df['fd'] * points_fd +
    stats_df['clutches'] * points_clutches +
    stats_df['aces'] * points_aces
)

# Display the updated DataFrame with total_points
stats_df
```

```python
# Find rows where player_name is 'Apoth'
p_stat = stats_df[stats_df['player_name'] == 'Apoth']
p_stat
```

```python
# Calculate total points per series per player
series_stats_df = stats_df.groupby(['series_id', 'player_name'])['total_points'].sum().reset_index()
```

```python
series_stats_df
```

```python
# Calculate total points per player across all series
total_points_per_player = series_stats_df.groupby('player_name')['total_points'].sum().reset_index()
# Sort by total points (descending order)
total_points_per_player_sorted = total_points_per_player.sort_values(by='total_points', ascending=False)
total_points_per_player_sorted
```
