### Links

```python
# Function to read and print links from the file
def read_links(file_path):
    with open(file_path, 'r') as file:
        links = file.readlines()
        links = [link.strip() for link in links]
    return links

file_path = '../Data/perf_list.txt'
# Access the links from the file
perf_list = read_links(file_path)
print(f"Links read from file {file_path}")

file_path = '../Data/over_list.txt'
# Access the links from the file
over_list = read_links(file_path)
print(f"Links read from file {file_path}")
```

# Merge DFS

```python
len_perf = len(perf_df)

# Select only the rows in over_df that are within the range of len_perf
over_df = over_df.iloc[:len_perf]
# Select only the rows in agent_df that are within the range of len_perf
agent_df = agent_df.iloc[:len_perf]
```

```python
# Perform the merge
agent_over_merge = pd.merge(agent_df, over_df, on=['game_id', 'player_name'], how='left')
```

```python
agent_over_merge.head()
```

```python
# Perform the merge
stats_agent_merge = pd.merge(agent_over_merge, perf_df, on=['game_id', 'player_name'], how='left')
```

```python
stats_agent_merge.head()
```

```python
# Merge the two DataFrames on the 'game_id' column
merged_df = pd.merge(stats_agent_merge, scores_df, on='game_id')
merged_df
```

```python
# Calculate the score differential
merged_df['score_dif'] = merged_df.apply(
    lambda row: row['home_score'] - row['away_score'] if row['team_abrev'] == row['home_team'] 
    else row['away_score'] - row['home_score'], axis=1)

# Calculate the total rounds played
merged_df['round_total'] = merged_df['home_score'] + merged_df['away_score']

# Drop unnecessary columns
# merged_df = merged_df.drop(columns=['home_team', 'away_team', 'home_score', 'away_score'])

new_column_order = ['game_id', 'map_name', 'map_duration', 'player_name', 'agent', 'team_abrev', 'kills', 'deaths', 'assists',
                   'adr', 'fk', 'fd', 'clutches', 'aces', 'score_dif', 'round_total']
    
# Reorder the columns
merged_df = merged_df[new_column_order]
```
