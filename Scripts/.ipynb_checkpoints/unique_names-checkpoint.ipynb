```python
# import dependencies
import pandas as pd
```

```python
file_path = '../Data/over_df.csv'
over_df = pd.read_csv(file_path)
over_df.head(5)
```

```python
# Group by player_name and aggregate team_abrev
unique_players = over_df.groupby('player_name')['team_abrev'].agg(lambda x: x.unique()[0]).reset_index()
```

```python
unique_players
```

```python
# Save DataFrame to CSV file
unique_players.to_csv('../Data/player_data.csv', index=False)
```
